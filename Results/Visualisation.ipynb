```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

totdosx = [item for sublist in np.genfromtxt('/home/armin/Downloads/Exjobb/Basic_Tutorial/total_dos.dat')[:, :1] for item in sublist]
totdosy = [item for sublist in np.genfromtxt('/home/armin/Downloads/Exjobb/Basic_Tutorial/total_dos.dat')[:, 1:] for item in sublist]
y1 = [item for sublist in np.genfromtxt('/home/armin/Downloads/Exjobb/Basic_Tutorial/Rapport/1TDOS')[1:, :] for item in sublist]
y2 = [item for sublist in np.genfromtxt('/home/armin/Downloads/Exjobb/Basic_Tutorial/Rapport/2TDOS')[1:, :] for item in sublist]
y3 = [item for sublist in np.genfromtxt('/home/armin/Downloads/Exjobb/Basic_Tutorial/Rapport/3TDOS')[1:, :] for item in sublist]
y4 = [item for sublist in np.genfromtxt('/home/armin/Downloads/Exjobb/Basic_Tutorial/Rapport/utTDOS')[1:, :] for item in sublist]
df = pd.DataFrame({'x': totdosx, 'STDEV_1E-2': y1, 'STDEV_1E-1': y2, 'MODULATED': y3, 'OUTCAR DB': y4, 'DFT(VASP)': totdosy})
fig = go.Figure()
line_styles = ['solid', 'dot', 'dash', 'longdash', 'dashdot']
entry_buttons = [{
    'label': 'Show All',
    'method': 'update',
    'args': [
        {'visible': [True] * len(df.columns[1:])},
        {'title': 'Phonon Density of States'}
    ]
}]
line_style_buttons = []
for i, column in enumerate(df.columns[1:]):
    entry_button = {
        'label': column,
        'method': 'update',
        'args': [
            {'visible': [True if j == i else False for j in range(len(df.columns[1:]))]},
            {'title': column}
        ]
    }
    entry_buttons.append(entry_button)
    line_style_button = {
        'label': line_styles[i % len(line_styles)].capitalize(),
        'method': 'update',
        'args': [
            {'line.dash': line_styles[i % len(line_styles)]},
            {}
        ]
    }
    line_style_buttons.append(line_style_button)
for i, column in enumerate(df.columns[1:]):
    fig.add_trace(go.Scatter(x=df['x'], y=df[column], mode='lines', name=column, line=dict(dash='solid')))

updatemenu = [
    {
        'buttons': entry_buttons,
        'direction': 'down',
        'showactive': True,
        'x': 0,
        'y': 1.35,
    },
    {
        'buttons': line_style_buttons,
        'direction': 'down',
        'showactive': True,
        'x': 0,
        'y': 1.2,
    }
]
fig.update_layout(
    title={'text': 'Phonon Density of States', 'x': 0.5, 'y': 0.9, 'font': {'size': 36}},
    xaxis_title='Frequency (THz)',
    yaxis_title='Density of States',
    updatemenus=updatemenu,
    showlegend=True
)
fig.show()
```
